```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("../Data/dataset_satelit.csv")
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
df['Mg'] = pd.to_numeric(df['Mg'], errors='coerce')
df = df.dropna()
```

```python
X_3 = df[['b2', 'b3', 'b4', 'b8', 'b11', 'Sigma_VV', 'Sigma_VH']]
y_3 = df['N']
print(df.columns.tolist())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_3, y_3, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'R2: {r2}')
print(f'RMSE: {rmse}')
```

```python
coeff = pd.DataFrame({
    'Fitur': X_3.columns,
    'Koefisien': model.coef_
})
print(coeff)
```

```python
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
model_sm = sm.OLS(y_train, X_train_sm).fit()
print(model_sm.summary())
```
